```python
'''

Given an array of integers, implement a linked list out of the array keys. The solution should create a new node for every key and insert it onto the list's front.

Input : [1, 2, 3, 4, 5]
Output: 1 —> 2 —> 3 —> 4 —> 5 —> None

'''

class Node:
    def __init__(self, data=None, next=None):
        self.data = data	# data field
        self.next = next	# pointer to the next node


class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def construct(self, keys: List[int]) -> Node:
        if not keys:
            return Node()
        head = None
        n = len(keys)
        i = n - 1
        while i >= 0:
            curr_node = Node(keys[i], None)
            curr_node.next = head
            head = curr_node
            i-= 1        
        return head

```

```python
'''

Given a singly-linked list of integers, which may contain a cycle, detect and report the cycle if present.

Input: 1 → 2 → 3 → 4 ─┐
		   └──────────┘
Output: True

Input: 1 → 2 → 3 → 4 → None
Output: False

'''



class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def detectCycle(self, head: Node) -> bool:
        if not head:
            return False       

        curr = head 
        while curr:
            if getattr(curr, "visited", False):
                return True
            curr.visited = True
            curr = curr.next
        
        return False
```

```python
'''

Given a singly-linked list of integers, split it into two sublists – one for the front half and one for the back half. The solution should return a tuple of the front half and the back half.

Input: 2 —> 3 —> 5 —> 7 —> None
Output: ([2 —> 3 —> None], [5 —> 7 —> None])

If the total number of elements in the list is odd, the extra element should go in the front list.

Input: 2 —> 3 —> 5 —> 7 —> 9 —> None
Output: ([2 —> 3 —> 5 —> None], [7 —> 9 —> None])

'''



class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def frontBackSplit(self, head: Node) -> Tuple[Node]:
        if not head:
            return tuple([None, None])
        
        s = []
        
        curr = head
        
        while curr:
            s.append(curr)
            curr = curr.next

        n = len(s)
        
        middle = (n - 1) // 2
        
        n1 = s[middle]
        
        next = n1.next 
        
        n1.next = None
        
        res = (head, next)
        
        return res
```

```python
'''

Given an unsorted linked list of integers, remove duplicate nodes from it by traversing the list only once. The solution should preserve the order of elements appearing in the list.

Input: 5 —> 3 —> 4 —> 2 —> 5 —> 4 —> 1 —> 3 —> None
Output: 5 —> 3 —> 4 —> 2 —> 1 —> None

'''


class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def removeDuplicates(self, head: Node) -> Node:
        
        if not head or not head.next:
            return head
        
        prev = head
        curr = prev.next
        
        s = set()
        s.add(prev.data)
        while curr:
            if curr.data not in s:
                prev.next = curr
                prev = curr
                s.add(curr.data)
    
            curr = curr.next
            
        prev.next = None
        
        return head

```

```python
'''

Given two sorted singly-linked lists of integers, return a new list representing their intersection.

Input:

X: 1 —> 4 —> 7 —> 10 —> None
Y: 2 —> 4 —> 6 —> 8 —> 10 —> None

Output: 4 —> 10 —> None

'''


class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def sortedIntersect(self, X: Node, Y: Node) -> Node:
        if not X:
            return None
        if not Y:
            return None
        x, y = X, Y
        
        nh = Node()
        res = nh
        while x and y:
            if x.data < y.data:
                x = x.next
            elif x.data > y.data:
                y = y.next
            else:
                nh.next = Node(x.data, None)
                nh = nh.next
                x = x.next
                y = y.next
        return res.next

```

```python
'''

Given a singly-linked list of strings, check whether the concatenation of all values in the list together forms a palindrome. It is not permissible to construct a string out of the linked list nodes and check that string for palindrome.

Input: 'AA' —> 'XYZ' —> 'CD' —> 'C' —> 'ZYX' —> 'AA' —> None
Output: True
Explanation: The string 'AAXYZCDCZYXAA' is palindrome

Input: 'A' —> 'B' —> 'C' —> 'DC' —> 'B' —> None
Output: False
Explanation: The string 'ABCDCB' is not a palindrome


Note that Punctuation, capitalization, and spaces should not be ignored.

Input: 'step on' —> 'no pets' —> None
Output: True
Explanation: The string 'step onno pets' is a palindrome

'''


class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''
    def construct(self, head, x='', y=''):
        if head is None:        # base case
            return x, y

        x += head.data
        x, y = self.construct(head.next, x, y)
        y += head.data[::-1]
        return x, y

    def isPalindrome(self, head: Node) -> bool:
        # Write your code here...
        x, y = self.construct(head)

        # check if the linked list is palindromic
        return x == y


```
