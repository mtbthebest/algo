```python
'''

Given an integer array, find the maximum difference between two elements in it such that the smaller element appears before the larger element. If no such pair exists, return any negative number.

Input : [2, 7, 9, 5, 1, 3, 5]
Output: 7
Explanation: The pair with the maximum difference is (2, 9).

Input : [5, 4, 3, 2, 1]
Output: -1 (or any other negative number)
Explanation: No pair with the maximum difference exists.

'''


class Solution:
    def findMaxDiff(self, nums: List[int]) -> int:
        if len(nums) <= 1:
            return -1
        minimum, maximum = nums[0], nums[0]
        diff = -1
        for i in range(1, len(nums)):
            if nums[i] > maximum or (nums[i] - minimum > diff and \
                                     nums[i] - minimum > 0):
                maximum = nums[i]
                diff = maximum - minimum
            if nums[i] < minimum:
                minimum = nums[i]
            
            
        return diff
Solution().findMaxDiff([1, 1, 1, 1, 1])
```

```python
'''

Given two integer arrays, each of which is sorted in increasing order, merge them into a single array in increasing order, and return it.

Input: X = [1, 3, 5, 7], Y = [2, 4, 6]
Output: [1, 2, 3, 4, 5, 6, 7]

'''


class Solution:
    def merge(self, X: List[int], Y: List[int]) -> List[int]:
        # Write your code here...
        k = len(X) - 1
        if not X:
            return Y
        if not Y:
            return X
        for i in range(len(Y)):
            X.append(None)

        i = len(X) - 1
        j = len(Y) - 1
        while i >= 0:
            if j >= 0 and k >=0 and Y[j] >= X[k]:
                X[i] = Y[j]
                j -= 1
            elif k >= 0 and j >= 0 and Y[j] < X[k]:
                X[i] = X[k]
                k -= 1
            else:
                break
            i -= 1
        while i >= 0 and j >= 0:
            X[i] = Y[j]
            i -= 1
            j -= 1
        return X

    
Solution().merge([7], [5])

```

```python
'''

Given a positive number `N`, return an `N Ã— N` spiral matrix containing numbers from 1 to `N Ã— N` in a counterclockwise direction.

Input: N = 5

Output:
[
	[25, 24, 23, 22, 21],
	[10, 9,  8,  7,  20],
	[11, 2,  1,  6,  19],
	[12, 3,  4,  5,  18],
	[13, 14, 15, 16, 17]
]

'''


class Solution:
    
    def side_fill(self, x, first_el, row, left, right, reverse=False):
        start, stop, step = (left, right, 1) if not reverse else (right, left, -1)
        
        for i in range(start, stop, step):
            x[row][i] = first_el
            first_el -= 1
        
        return first_el
    
    def hill_fill(self, x, first_el, col, top, bottom, reverse=False):
        start, stop, step = (top, bottom, 1) if not reverse else (bottom, top, -1)        
        for i in range(start, stop, step):
            x[i][col] = first_el
            first_el -= 1
        
        return first_el
    
    
    def findSpiralOrder(self, N: int) -> List[List[int]]:
        # Write your code here...
        nsquare = N * N
        if nsquare == 0:
            return []
        if N == 1:
            return [[1]]
        m, n = N, N
        top, bottom, left, right = 0, m - 1, 0, n - 1
        first_el = nsquare
        x = [[None] * N for _ in range(N)]
        while top <= bottom and left <= right:
            if top == bottom and left == right:
                x[top][right] = first_el
                break
            
            for i in range(4):
                
                if i % 2 == 0:
                    if left >= right:
                        continue
                    if top == bottom and i == 2:
                        if first_el is not None:
                            x[top][right] == first_el
                            first_el -= 1
                        continue
                            
                            
                    first_el = self.side_fill(x, first_el, top if i == 0 else bottom,
                                              left, right, reverse=(i == 2))
                
                else:
                    if top >= bottom:
                        continue
                    if left == right and i == 3:
                        if first_el is not None:
                            x[bottom][right] == first_el
                            first_el -= 1
                        continue
                    first_el = self.hill_fill(x, first_el, left if i == 3 else right, 
                                              top, bottom, reverse=(i == 3))
            
            
            top += 1
            bottom -= 1
            right -= 1
            left += 1
        
        return x


Solution().findSpiralOrder(N=5)
```

```python
'''

Given a positive number, convert the number to the corresponding Excel column name.

Input: 1
Output: 'A'

Input: 27
Output: 'AA'

Input: 1001
Output: 'ALM'

Note: Excel supports upto 2^14 columns, i.e. 16,384 columns.

'''


class Solution:
    def getColumnName(self, n: int) -> str:
        # Write your code here...
        alphabet = {i: chr(i + 64) for i in range(1, 27)}
        res = []
        i = 1
        while n > 0:
            index = n % 26
            if index == 0:
                index = 26
                n = n // 26 -1
            else:
                n = n // 26
                
            char = alphabet[index]
            res.insert(0, char)
            
        return "".join(res)

    
Solution().getColumnName(52)
```

```python
'''

Given two strings, determine whether they are anagrams.

Any word that exactly reproduces the letters in another order is an anagram. In other words, X and Y are anagrams if by rearranging the letters of X, we can get Y using all the original letters of X exactly once.

Input: X = 'silent', Y = 'listen'
Output: True

Input: X = 'incest', Y = 'insect'
Output: True

'''


class Solution:
    def isAnagram(self, X: str, Y: str) -> bool:
        # Write your code here...
        xdict = dict()
        if not X and not Y:
            return True
        if not X:
            return False
        for k in X:
            if k not in xdict:
                xdict[k] = 0
            xdict[k] += 1

        for k in Y:
            if k in xdict:
                xdict[k] -= 1
                if xdict[k] == 0:
                    xdict.pop(k)
            else:
                return False
        return len(xdict) == 0

Solution().isAnagram(X = 'ab', Y = 'aab')
```

```python
'''

Given a sorted integer array and a target, determine if the target exists in the array in logarithmic time. If target exists in the array, the procedure should return the index of it.

Input: nums[] = [2, 3, 5, 7, 9], target = 7
Output: 3
Explanation: Element found at 4th (index 3) position

â€¢ If there are duplicate elements in the array, the procedure may return any index whose element is equal to the target.

Input: nums[] = [1, 2, 3, 4, 4, 5, 6, 7], target = 4
Output: 3 (or 4)
Explanation: Element found at the 4th (index 3) or 5th (index 4) position.

â€¢ If the target is not located, the procedure should return -1.

Input: nums[] = [1, 4, 5, 8, 9], target = 2
Output: -1

'''


class Solution:
    
    def binary_search(self, nums, low, high, target):
        if low == high:
            if nums[low] == target:
                return low
            return -1
        mid = low + (high - low) // 2
        
        if target > nums[mid]:
            return self.binary_search(nums, mid + 1, high, target)
        else:
            return self.binary_search(nums, low, mid, target)
    
    def findIndex(self, nums: List[int], target: int) -> int:
        # Write your code here...
        if not nums:
            return -1
        return self.binary_search(nums,0, len(nums) - 1, target)

    
Solution().findIndex(nums=  [1, 4, 5, 8, 9], target = 2)
```

```python
'''

Given an unsorted integer array, find the smallest missing positive integer in it.

Input: [1, 4, 2, -1, 6, 5]
Output: 3

Input: [1, 2, 3, 4]
Output: 5

'''

class Solution:
    """Solution: https://www.geeksforgeeks.org/find-the-smallest-positive-number-missing-from-an-unsorted-array/"""
    def findSmallestMissingNumber(self, nums: List[int]) -> int:
        # Segregate positive and negative
        j = 0
        n = len(nums)
        for i in range(n):
            if nums[i] <= 0:
                nums[i], nums[j] = nums[j], nums[i]
                j += 1
        
        if j == n:
            return 1
        nums = nums[j:]
        j = 0
        n = len(nums)
        for i in range(j, n):
            index = abs(nums[i])
            if index > 0 and index - 1 < n  and nums[index - 1] > 0:
                nums[index - 1] = -(nums[index - 1])
        for i in range(j, n):
            if nums[i] > 0:
                return i + 1
        
        return i + 2
                
        
    
        
Solution().findSmallestMissingNumber( [-1, 1, 0])
```

```python
'''

Given a nearly sorted array such that each of its elements may be misplaced by no more than one position from the correct sorted order, efficiently search a given element in it and return its index.

An element at index `i` in a correctly sorted order can be misplaced by the Â±1 position, i.e., it can be present at index `i-1` or `i` or `i+1` in the input array.

Input: nums[] = [2, 1, 3, 5, 4, 7, 6, 8, 9], target = 5
Output: 3
Explanation: Target 5 found at index 3

â€¢ If there are duplicate elements in the array, the procedure may return any index whose element is equal to the target.

Input: nums[] = [2, 1, 3, 5, 4, 5, 6, 8, 9], target = 5
Output: 3 (or 5)
Explanation: Target 5 found at index 3 and 5

â€¢ If the target is not located, the procedure should return -1.

Input: nums[] = [2, 1, 3, 5, 4, 7, 6, 8, 9], target = 10
Output: -1

'''


class Solution:
    def findIndex(self, nums: List[int], target: int) -> int:
        
        low, high  = 0, len(nums) - 1
        while low <= high:
            mid = (low + high) // 2
            
            if nums[mid] == target:
                return mid
            
            elif mid - 1 >= low and nums[mid - 1] == target:
                return mid - 1
            elif mid + 1 <= high and nums[mid + 1] == target:
                return mid + 1
            elif target > nums[mid]:
                low = mid + 2
            
            else:
                high = mid - 2
        
        
        return -1
        
```

```python
'''

Given an integer array `A`, find the peak element in it. An element `A[i]` is a peak element if it's greater than its neighbor(s). i.e.,

â€¢ A[i-1] <= A[i] >= A[i+1]	(for 0 < i < n-1)
â€¢ A[i-1] <= A[i]			(if i = n â€“ 1)
â€¢ A[i] >= A[i+1]			(if i = 0)


Input: [8, 9, 10, 12, 15]
Output: 15

Input: [10, 8, 6, 5, 3, 2]
Output: 10

â€¢ There might be multiple peak elements in an array, the solution should report any peak element.

Input: [8, 9, 10, 2, 5, 6]
Output: 10 or 6

Input: [8, 9, 2, 5, 6, 3]
Output: 9 or 6

â€¢ If the peak element is not located, the procedure should return -1.

Input: []
Output: -1

'''


class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        if not nums:
            return -1
    
        n = len(nums)
        if n == 1:
        	return nums[0]
        for i in range(n):
            if i == 0 and i + 1 < n and nums[i] >= nums[i+1]:
                return nums[i]
            elif i == n - 1 and i - 1 >= 0 and nums[i - 1] <= nums[i]:
                return nums[i]
            
            elif i - 1 > 0 and i + 1 < n and nums[i-1] <= nums[i] and nums[i] >= nums[i+1]:
                return nums[i]
        return -1
                
    
    
    
```

```python
'''

Given a sorted integer array `nums` and two positive integers `k` and `target`, find the `k` closest elements to `target` in the array. Here, `target` may or may not be present in the input array. The solution should return elements in the same order as present in the input array.

Input : nums[] = [10, 12, 15, 17, 18, 20, 25], k = 4, target = 16
Output: [12, 15, 17, 18]


â€¢ If `target` is less than or equal to the first element in the input array, the solution should return the first `k` elements.

Input : nums[] = [2, 3, 4, 5, 6, 7], k = 3, target = 1
Output: [2, 3, 4]


â€¢ If `target` is more than or equal to the last element in the input array, the solution should return the last `k` elements.

Input : nums[] = [2, 3, 4, 5, 6, 7], k = 2, target = 8
Output: [6, 7]


â€¢ In case multiple sets of `k` closest elements exists, the solution should return the first occuring set.

Input : nums[] = [10, 12, 15, 17, 20, 25], k = 3, target = 16
Output: [12, 15, 17]
Explanation: There are two k-closest element sets present in the array i.e., [12, 15, 17] and [15, 17, 20]. The solution returns set [12, 15, 17] as it appears before set [15, 17, 20] in the array.

'''


class Solution:
    def kClosestElements(self, nums: List[int], k: int, target: int) -> List[int]:
        if not nums:
            return []
        if k > len(nums):
            return []
        
        if target <= nums[0]:
            return nums[:k]
        if target >= nums[-1]:
            return nums[len(nums) - k:]
        n = len(nums)
        low, high = 0, n - 1
        index = None
        while (high - low) > 1:
            mid = (low + high) // 2
            if target == nums[mid]:
                index = mid
                break
            if target < nums[mid]:
                high = mid
            else:
                low = mid
        els = []
        i = index
        if i is None:
            i = low
            j = high
        else:
            j = min(i+1, n-1)
        
        while k > 0 and i >=0 and j < n:
            if abs(nums[i] - target) <= abs(nums[j] - target):
                els.insert(0, nums[i])
                i -= 1
            else:
                els.append(nums[j])
                j += 1
            k -= 1
        
        while k > 0:
            if i >= 0:
                els.insert(0, nums[i])
                i -= 1
            else:
                els.append(nums[j])
                j += 1
            
            k -= 1
                
        
        return els
            
            
Solution().kClosestElements([-10, -7, -6, 1, 5], k = 1, target = -4)
```

```python
'''

Given a set of strings, find their longest common prefix (LCP).

Input: ['technique', 'technician', 'technology', 'technical']
Output: 'techn'

Input: ['techie delight', 'tech', 'techie', 'technology', 'technical']
Output: 'tech'

'''


class Solution:
    def findLCP(self, words: List[str]) -> str:
        if not words:
            return ""
        
        lcp = words[0]
        for word in words[1:]:
            if lcp != word[:len(lcp)]:
                k = 0
                try:
                    while lcp[k] == word[k]:
                        k += 1
                except:
                    pass
                lcp = lcp[:k]
        return lcp
    
Solution().findLCP(['techie delight', 'tech', 'techie', 'technology', 'technical'])
```
