```python
'''

Given an integer array, replace each element with the product of every other element without using the division operator.

Input : [1, 2, 3, 4, 5]
Output: [120, 60, 40, 30, 24]

Input : [5, 3, 4, 2, 6, 8]
Output: [1152, 1920, 1440, 2880, 960, 720]

'''


class Solution:
    def get_right(self, i, right, nums):
        if right[i] is not None:
            return right[i]
        if i == len(right) - 1:
            right[i] = 1
        else:
            right[i] = nums[i + 1] * self.get_right(i+1, right, nums)

        return right[i]

    def get_left(self, i, left, nums):
        if left[i] is not None:
            return left[i]
        if i == 0:
            left[i] = 1
        else:
            left[i] = nums[i-1] * self.get_left(i-1, left, nums)

        return left[i]

    def rearrange(self, nums: List[int]) -> None:
        if not nums:
            return
        left, right = [None] * len(nums), [None] * len(nums)
        self.get_right(0, right, nums)
        self.get_left(len(left) - 1, left, nums)
        print(right)
        print(left)
        for i in range(len(nums)):
            nums[i] = self.get_left(i, left, nums) * self.get_right(i, right, nums)
        print(nums)


Solution().rearrange([1, 2, 3, 4, 5])
```

```python
'''
# TODO the problem did not specify from 0 to n-1 descending allowed and from 0 to n-1 ascending allowed


The Longest Bitonic Subarray (LBS) problem is to find a contiguous subarray of a given sequence in which the subarrayâ€™s elements are first sorted in increasing order, then in decreasing order, and the subarray is as long as possible.

Input : [3, 5, 8, 4, 5, 9, 10, 8, 5, 3, 4]
Output: [4, 5, 9, 10, 8, 5, 3]

In case the multiple bitonic subarrays of maximum length exists, the solution can return any one of them.

Input : [-5, -1, 0, 2, 1, 6, 5, 4, 2]
Output: [-5, -1, 0, 2, 1] or [1, 6, 5, 4, 2]

'''


class Solution:
    def merge(self, nums, low, mid, high):
        print(low, mid, high)
        if nums[mid] == 1:
            # Find k
            k = mid
            while k <= high and k < len(nums) and nums[k] == 1:
                k += 1
            if k == high + 1:
                if high + 1 != len(nums):
                    return slice(0, 0)
            j = k
            k -= 1
            # Find i
            i = mid
            while i >= low and i >= 0 and nums[i] == 1:
                i -= 1
            if i < 0:
                i = 0

            if nums[i] == -1:
                i += 1

            i = max(low, i)

            # Find j
            while j <= high and j < len(nums) and nums[j] == -1:
                j += 1

            if j == high + 1:
                j -= 1
        else:
            # nums[mid] == -1
            j = mid
            k = mid
            while k >= 0 and k > low and nums[k] == -1:
                k -= 1
            if k == low and nums[k] == -1:
                if low != 0:
                    return slice(0, 0)
            i = k
            if nums[i] == -1 and low == 0:
                while i >= low and nums[i] == 1:
                    i -= 1
            while j <= high and nums[j] == -1:
                j += 1

            if j == high + 1:
                j -= 1

        return slice(i, j + 1)

    def mergeSort(self, nums, low, high):
        if low == high:
            return slice(0, 0)

        mid = low + (high - low) // 2

        if mid - low + 1 > 2:
            left = self.mergeSort(nums, low, mid)
        else:
            left = slice(0, 0)

        if high - mid > 2:
            right = self.mergeSort(nums, mid+1, high)
        else:
            right = slice(0, 0)

        middle = self.merge(nums, low, mid, high)
        
        left_length, right_length, middle_length = left.stop - left.start, right.stop - right.start, \
            middle.stop - middle.start
        
        if left_length >= right_length and left_length >= middle_length:
            return left
        
        elif right_length > max(left_length, middle_length):
            return right
        else:
            return middle

    def findBitonicSubarray(self, nums: List[int]) -> List[int]:
        if not nums:
            return []
        snums = []
        for i in range(len(nums) - 1):
            snums.append(1 if nums[i+1] - nums[i] >= 0 else -1)
        snums.append(snums[-1])

        monotic_value = snums[0]

        is_monotic = True
        for el in snums:
            if el == -monotic_value:
                is_monotic = False
                break
        if is_monotic:
            return nums
        print(snums)
        indice = self.mergeSort(snums, 0, len(nums) - 1)
        return nums[indice]


# Solution().findBitonicSubarray([3, 5, 8, 4, 5, 9, 10, 8, 5, 3, 4])
# Solution().findBitonicSubarray([-5, -1, 0, 2, 1, 6, 5, 4, 2])
# Solution().findBitonicSubarray([5, 4, 3, 2, 1])
Solution().findBitonicSubarray([5, 1, 0, -2, -1, -6, -5, -8, -7])
```

```python
'''

Given an integer array, find the maximum difference between two elements in it such that the smaller element appears before the larger element. If no such pair exists, return any negative number.

Input : [2, 7, 9, 5, 1, 3, 5]
Output: 7
Explanation: The pair with the maximum difference is (2, 9).

Input : [5, 4, 3, 2, 1]
Output: -1 (or any other negative number)
Explanation: No pair with the maximum difference exists.

'''


class Solution:
    def findMaxDiff(self, nums: List[int]) -> int:
        if len(nums) <= 1:
            return -1
        minimum, maximum = nums[0], nums[0]
        diff = -1
        for i in range(1, len(nums)):
            if nums[i] > maximum or (nums[i] - minimum > diff and \
                                     nums[i] - minimum > 0):
                maximum = nums[i]
                diff = maximum - minimum
            if nums[i] < minimum:
                minimum = nums[i]
            
            
        return diff
Solution().findMaxDiff([1, 1, 1, 1, 1])
```
