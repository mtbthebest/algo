```python
'''

Given an array of integers, implement a linked list out of the array keys. The solution should create a new node for every key and insert it onto the list's front.

Input : [1, 2, 3, 4, 5]
Output: 1 —> 2 —> 3 —> 4 —> 5 —> None

'''

class Node:
    def __init__(self, data=None, next=None):
        self.data = data	# data field
        self.next = next	# pointer to the next node


class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def construct(self, keys: List[int]) -> Node:
        if not keys:
            return Node()
        head = None
        n = len(keys)
        i = n - 1
        while i >= 0:
            curr_node = Node(keys[i], None)
            curr_node.next = head
            head = curr_node
            i-= 1        
        return head

```

```python
'''

Given a singly-linked list of integers, which may contain a cycle, detect and report the cycle if present.

Input: 1 → 2 → 3 → 4 ─┐
		   └──────────┘
Output: True

Input: 1 → 2 → 3 → 4 → None
Output: False

'''



class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def detectCycle(self, head: Node) -> bool:
        if not head:
            return False       

        curr = head 
        while curr:
            if getattr(curr, "visited", False):
                return True
            curr.visited = True
            curr = curr.next
        
        return False
```

```python
'''

Given a singly-linked list of integers, split it into two sublists – one for the front half and one for the back half. The solution should return a tuple of the front half and the back half.

Input: 2 —> 3 —> 5 —> 7 —> None
Output: ([2 —> 3 —> None], [5 —> 7 —> None])

If the total number of elements in the list is odd, the extra element should go in the front list.

Input: 2 —> 3 —> 5 —> 7 —> 9 —> None
Output: ([2 —> 3 —> 5 —> None], [7 —> 9 —> None])

'''



class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def frontBackSplit(self, head: Node) -> Tuple[Node]:
        if not head:
            return tuple([None, None])
        
        s = []
        
        curr = head
        
        while curr:
            s.append(curr)
            curr = curr.next

        n = len(s)
        
        middle = (n - 1) // 2
        
        n1 = s[middle]
        
        next = n1.next 
        
        n1.next = None
        
        res = (head, next)
        
        return res
```

```python
'''

Given an unsorted linked list of integers, remove duplicate nodes from it by traversing the list only once. The solution should preserve the order of elements appearing in the list.

Input: 5 —> 3 —> 4 —> 2 —> 5 —> 4 —> 1 —> 3 —> None
Output: 5 —> 3 —> 4 —> 2 —> 1 —> None

'''


class Solution:

    '''
    A singly-linked list node is defined as:

    class Node:
        def __init__(self, data=None, next=None):
            self.data = data	# data field
            self.next = next	# pointer to the next node
    '''

    def removeDuplicates(self, head: Node) -> Node:
        
        if not head or not head.next:
            return head
        
        prev = head
        curr = prev.next
        
        s = set()
        s.add(prev.data)
        while curr:
            if curr.data not in s:
                prev.next = curr
                prev = curr
                s.add(curr.data)
    
            curr = curr.next
            
        prev.next = None
        
        return head

```

```python
'''

Given the root of a binary tree, return the level order traversal of its nodes' values. The solution should consider the binary tree nodes level by level from left to right, i.e., process all nodes of level 1 first, followed by all nodes of level 2, and so on.

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  /		  / \
	 /	  	 /	 \
	4		5	  6
		   / \
		  /   \
		 7	   8

Output: [1, 2, 3, 4, 5, 6, 7, 8]

'''


class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def findLevelOrderTraversal(self, root: Node) -> List[int]:
        if not root:
            return []
        
        q = deque()
        res = []
        q.append(root)
        while q:
            node = q.popleft()
            res.append(node.data)
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
        return res

```

```python
'''

Given the root of two binary trees, x and y, check if x is identical to y. Two binary trees are identical if they have identical structure and their contents are also the same.

Input:
		   1						1
		 /   \					  /   \
		/	  \					 /	   \
	   2	   3				2		3
	  /	\	  / \			   / \	   / \
	 /	 \ 	 /	 \			  /	  \	  /	  \
	4	  5	6	  7			 4	   5 6	   7

Output: True
Explanation: Both binary trees have the same structure and contents.

Input:
		   1						1
		 /   \					  /   \
		/	  \					 /	   \
	   2	   3				2		3
	  /	\	  / \			   / \	   /
	 /	 \ 	 /	 \			  /	  \	  /
	4	  5	6	  7			 4	   5 6

Output: False
Explanation: Both binary trees have different structures.

Input:
		   1						1
		 /   \					  /   \
		/	  \					 /	   \
	   2	   3				2		3
	  /	\	  / \			   / \	   / \
	 /	 \ 	 /	 \			  /	  \	  /	  \
	4	  5	6	  7			 4	   5 6	   8

Output: False
Explanation: Both binary trees have the same structure but differ in nodes' values.

'''


class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def isIdentical(self, x: Node, y: Node) -> bool:
        if not x and not y:
            return True
        if (x and not y) or (y and not x):
            return False
        
        qx, qy = deque(), deque()
        
        qx.append(x)
        qy.append(y)
        while qx and qy:
            
            nx = qx.pop()
            ny = qy.pop()
            
            if nx.data != ny.data:
                return False
            
            if nx.left:
                qx.appendleft(nx.left)
            if nx.right:
                qx.append(nx.right)
            
            if ny.left:
                qy.appendleft(ny.left)
            if ny.right:
                qy.append(ny.right)
                
            
        return not qx and not qy

```

```python
'''

Given the root of a binary tree, return the binary tree's height. The height of the binary tree is the total number of edges or nodes on the longest path from the root node to the leaf node.

The solution should consider the total number of nodes in the longest path. For example, an empty tree's height is 0, and the tree's height with only one node is 1.

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  /	\	  / \
	 /	 \ 	 /	 \
	4	  5	6	  7

Output: 3


Input:
		   1
		  /
		 /
		2
	   /
	  /
	 3
	/
   /
  4

Output: 4

'''


class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def findHeight(self, root: Node) -> int:
        if not root:
            return 0
        
        height = 1
        
        q = deque()
        q.append((root, 1))
        
        while q:
            node, level = q.pop()
            
            height = max(level, height)
            
            left, right = node.left, node.right
            
            if left:
                q.appendleft((left, level + 1))
            if right:
                q.appendleft((right, level + 1))
                
        
        return height
```

```python
'''

Given the root of a binary tree with leaf nodes forming a circular doubly linked list, return the binary tree's height. Note that the leaf node's left and right pointers will act as a previous and next pointer of the circular doubly linked list, respectively.

The height of the binary tree is the total number of edges or nodes on the longest path from the root node to the leaf node. The solution should consider the total number of nodes in the longest path. For example, an empty tree's height is 0, and the tree's height with only one node is 1.

Input: Below binary tree (The leaf nodes [7, 5, 6] are connected using the left and right pointers and forming a circular doubly linked list)

		   1
		 /   \
		/	  \
	   2	   3
	  /	\		\
	 /	 \ 	 	 \
	4	  5		  6
   /
  /
 7

Output: 4

'''

class Node:
    def __init__(self, data=None, left=None, right=None):
        self.data = data	# data field
        self.left = left	# pointer to the left child
        self.right = right	# pointer to the right child

class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def findHeight(self, root: Node) -> int:
        if root is None:
            return 0

        if ((root.left and root.left.right == root) and
                    (root.right and root.right.left == root)):
            return 1

        # recur for the left and right subtree and consider maximum depth
        return 1 + max(self.findHeight(root.left), self.findHeight(root.right))

```

```python
'''

Given the root of a binary tree and two positive numbers m and n where m >= n, return values of all nodes between level m and level n. The nodes for each level should be processed from left and right.

Input: Below binary tree, m = 2, n = 3

		   1
		 /   \
		/	  \
	   2	   3
			  / \
			 /	 \
			4	  5
		   / \	   \
		  /   \		\
		 6	   7	 8

Output: [2, 3, 4, 5]

Note: If n is more than the number of levels in the binary tree, the solution return nodes till last level. For example, if the starting level is 2 and the ending level is 7, the solution should return [2, 3, 4, 5, 6, 7, 8] for above binary tree.

'''



class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def findNodes(self, root: Node, m: int, n: int) -> List[int]:
        if not root:
            return []
        
        lookup = defaultdict(deque)
        q = deque()
        q.append((root,  1, None))
        while q:
            node, level, parent = q.pop()
            if not node:
                continue
            lookup[level].append(node.data)
            left, right = node.left, node.right
            q.appendleft((left, level + 1, node))
            q.appendleft((right, level + 1, node))
        res = []
        for i in range(m, n + 1):
            if i in lookup:
                res.extend(lookup[i])
        return res

```

```python
'''

Given the root of a binary tree and two tree nodes, x and y, return the distance between x and y in the binary tree. The distance between two nodes is defined as the total number of edges in the shortest path from one node to other.

For example, consider the following binary tree.

		  1
		/   \
	   /	 \
	  2		  3
	   \	 / \
		4   5   6
		   /	 \
		  7		  8

Input: x = Node 7, y = Node 6
Output: 3

Input: x = Node 4, y = Node 8
Output: 5

Input: x = Node 5, y = Node 5
Output: 0

Note: The solution should return a negative value if either x or y is not the actual node in the tree.

'''



class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def findDistance(self, root: Node, x: Node, y: Node) -> int:
        if not root or not x or not y:
            return -1
        if x is y:
            return 0
    
        lookup = {}
        q = deque()
        q.append((root,  1, None))
        while q:
            node, level, parent = q.pop()
            if not node:
                continue
            lookup[node] = (level, parent)
            left, right = node.left, node.right
            q.appendleft((left, level + 1, node))
            q.appendleft((right, level + 1, node))
        if x not in lookup or y not in lookup:
            return -1

        u, v = (x , y) if lookup[x][0] <= lookup[y][0] else (y, x)
        ulev, vlev = lookup[u][0], lookup[v][0]
        
        dst = 0
        while ulev < vlev:
            v = lookup[v][1]
            vlev -= 1
            dst += 1 
        if u is v:
            return dst
        
        while ulev > 0:
            u, v = lookup[u][1], lookup[v][1]
            dst += 2
            if u is v:
                return dst
            ulev -= 1
        
        return dst
```

```python
'''

Given the root of a binary tree, return the difference between the sum of all nodes present at odd levels and the sum of all nodes present at even level.

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  /		  / \
	 /	  	 /	 \
	4	  	5	  6
		   / \
		  /	  \
		 7	   8

Output: -4

Explanation: The difference is (1 + 4 + 5 + 6) - (2 + 3 + 7 + 8) = -4

'''



class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''
    
    def recurse(self, root, level, even, odd):
        if not root:
            return even, odd
        even, odd = self.recurse(root.left, level + 1, even, odd)
        even, odd = self.recurse(root.right, level + 1, even, odd)
        
        if level % 2:
            odd += root.data
        else:
            even += root.data
        
        return even, odd 
        

    def findDifference(self, root: Node) -> int:
        if not root:
            return 0
        
        even, odd = self.recurse(root, 1, 0, 0)
        
        return odd - even

```

```python
'''

Given the root of a binary tree, return the maximum sum of a path between any two leaves in the binary tree. You may assume that the binary tree is not skewed and contains at-least two nodes.

Input:

		  1
		/   \
	   /	 \
	  2		  3
	   \	 / \
	   -4   5   6
		   / \
		  7   8

Output: 22

Explanation: The maximum sum path between two leaves is [8, 5, 3, 6].

'''


    
class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''
    
    def recurse(self, root):
        if not root:
            return [], None

        if not root.left and not root.right:
            return [(root.data, )], None
            
        left_leaves, left_best = self.recurse(root.left)
        right_leaves, right_best = self.recurse(root.right)
        
        if (left_best and not right_best) or (left_best and right_best and sum(left_best) > sum(right_best)):
            best = left_best
        else:
            best = right_best
        leaves = []
        
        for path in (left_leaves + right_leaves):
            leaves.append(path + (root.data,))
        if not right_leaves or not left_leaves:
            return leaves, best
        
        if best is None:
            best = [-sys.maxsize]

        for left_path in left_leaves:
            for right_path in right_leaves:
                total_sum = sum(left_path) + sum(right_path) + root.data
                if total_sum > sum(best) :
                    best = left_path + (root.data, ) + right_path
        return leaves, best

    def findMaximumSum(self, root: Node) -> int:
        if not root:
            return 0
        leaves, best = self.recurse(root)
        if best is None:
            return 0
        return sum(best)
```

```python
'''
# TODO
Given the root of a non-empty binary tree, return the maximum path sum between any two nodes in the binary tree. The path can start and end at any node in the tree and need not go through the root.

Input:

			 1
		   /   \
		 /		 \
		2		  10
	   / \		 / \
	  /	  \		/	\
	 -1	  -4   -5	 -6
		  /	  / \
		 /	 /	 \
		3	7	  4
			 \
			  \
			  -2

Output: 15

Explanation: The maximum sum path is [2, 1, 10, -5, 7].

'''

CNode = namedtuple('CNode', ['left', 'right', 'best'])

class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''
    
    def trace(self, root):
        if not root:
            return None
        
        if not root.left and not root.right:
            return None
        
        left_cnode: CNode = self.trace(root.left)
        right_cnode: CNode = self.trace(root.right)
        
        if not left_cnode and not right_cnode:
            if not root.left:
                left_sum = -sys.maxsize
            else:
                left_sum = max(-sys.maxsize,  root.data + root.left.data)
            
            if not root.right:
                right_sum = -sys.maxsize
            else:
                right_sum = max(-sys.maxsize,  root.data + root.right.data)
            
            if root.left and root.right:
                best = max(left_sum, right_sum, left_sum + right_sum  - root.data) 
            
            else:
                best = max(left_sum, right_sum, left_sum + right_sum)
            
                
            return CNode(left=left_sum, right=right_sum, best=best)

        if not left_cnode:
            right_sum = max(right_cnode.left + root.data, right_cnode.right + root.data, root.data + root.right.data )
            return CNode(left=-sys.maxsize, 
                         right=right_sum,
                         best=max(right_cnode.best, right_sum))
        if not right_cnode:
            left_sum = max(left_cnode.left + root.data, left_cnode.right + root.data, root.data + root.left.data )
            return CNode(right=-sys.maxsize, 
                         left=left_sum,
                         best=max(left_cnode.best, left_sum))
        if left_cnode:
            left_sum = max(left_cnode.left + root.data, left_cnode.right + root.data, root.data + root.left.data )
        else:
            left_sum = root.data + root.left.data
        
        if right_cnode:
            
            right_sum = max(right_cnode.left + root.data, right_cnode.right + root.data, root.data + root.right.data )
        else:
            right_sum = root.data + root.right.data


        best = max((left_cnode.best if left_cnode else -sys.maxsize), 
                   (right_cnode.best if right_cnode else -sys.maxsize),
                   left_sum, 
                   right_sum, 
                   left_sum + right_sum - root.data)
        
        

        return CNode(left_sum, right_sum, best) 
        
    def findMaximumSum(self, root: Node) -> int:
        if not root:
            return 0
        
        res = self.trace(root)
        if not res:
            return 0
        return res.best

```

```python
'''
Given the root of a binary tree, convert it into a Left–child right–sibling (LC–RS) binary tree.

Each node in the LC–RS binary tree has two pointers: one to the node's left child and one to its next sibling in the original binary tree. So starting with the root, each node's leftmost child in the original tree is made its left child in the LC–RS binary tree, and its nearest sibling to the right in the original tree is made its right child in the binary tree.

Input:

			 1
		   /   \
		 /		 \
		2		  3
	  /  \		 /
	 /	  \		/
	4	   5   6
			  / \
			 /	 \
			7	  8

Output:

			1
		  /
		/
	   2
	 /   \
   /	   \
  4			3
   \	   /
	\	  /
	 5   6
		/
	   /
	  7
	   \
		\
		 8

Explanation:

Rewrite the binary tree shown by putting the left child node to one level below its parents and by placing the sibling next to the left child at the same level.

			 1
		   /
		 /
		2 ------- 3
	  /  \
	 /	  \
	4 ---- 5 -- 6
			   /
			  /
			 7 --- 8

Then transform this tree into a LC-RS binary tree by turning each sibling 45° clockwise.

'''


class Solution:

	def buildLCRSBinaryTree(self, root: Node) -> None:
		if not root:
			return root

		left, right = root.left, root.right

		left = self.buildLCRSBinaryTree(root.left)
		right = self.buildLCRSBinaryTree(root.right)

		if left is not None:
			left.right = right


		root.left = left if root.left else right
		root.right = None
	
		return root
		
```

```python
'''

Given a postfix expression, construct a binary expression tree from it and return its root. The binary expression tree is a binary tree whose leaves are operands, such as constants or variable names, and the other nodes contain operators.

Input: ab+cde+**
Output: Root of the following expression tree.

			 *
		   /   \
		 /		 \
		+		  *
	   / \		 / \
	  /	  \		/	\
	 a	   b   c	 +
					/ \
				   /   \
				  d		e

'''



class Solution:
 
	def constructExpressionTree(self, postfix: str) -> Node:
		if not postfix:
			return None
		ops = set(["+", "*", "-", "/", "%"])
		stack = []
		root = node = Node(postfix[0])
		for i in range(len(postfix)):
			el = postfix[i]
			root = node = Node(el)
			if el in ops:
				a, b = stack[-2:]
				stack.pop()
				stack.pop()
				node.left = a
				node.right = b
			stack.append(node)
		return root
```

```python
'''

Given the root of a binary expression tree representing algebraic expressions, evaluate it and return its value. A binary expression tree is a binary tree, where the operators are stored in the tree's internal nodes, and the leaves contain constants.

Assume that each node of the binary expression tree has zero or two children. The supported operators are +(addition), −(subtraction), *(multiplication), ÷(division) and ^(exponentiation).

Input:

			(+)
		   /   \
		 /		 \
	   (*)		 (/)
	   / \		 / \
	  /	  \		/	\
	(-)	   5   21	 7
	/ \
   /   \
  10	5

Output: 28

Explanation: The corresponding infix notation is ((10-5)*5)+(21/7) = 28 which can be produced by traversing the expression tree in an inorder fashion.

'''


class Solution:

    '''
    An expression tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def evaluate(self, root: Node) -> float:
        if not root:
            return 0
        ops = set(["+", "*", "-", "/", "%"])
        
        if root.data in ops:
            left = self.evaluate(root.left) 
            right = self.evaluate(root.right)
            op = root.data
            return eval(f"{left}{op}{right}")
        return float(root.data)
                    
```

```python
'''

Given the root of a binary tree, check if the binary tree has a symmetric structure or not, i.e., left and right subtree mirror each other.

Input:
		   1
		 /   \
		/	  \
	   2	   3
		\	  /
		 \	 /
		  4	5

Output: True

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  / \	  / \
	 /   \	 /	 \
	7	  8	5	  6

Output: True

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  /		  /
	 /		 /
	7		5

Output: False

'''


class Solution:
	def recurse_left(self, root, arr, level):
		if not root:
			return level
		arr.append((root.left, level + 1))
		arr.append((root.right, level + 1))

		left_level = self.recurse_left(root.left, arr, (level + 1))
		right_level = self.recurse_left(root.right, arr, (level + 1))

		return max(left_level, right_level)
  
	def recurse_right(self, root, arr, level):
		if not root:
			return level
		arr.append((root.right, level + 1))
		arr.append((root.left, level + 1))

		right_level = self.recurse_right(root.right, arr, (level + 1))
		left_level = self.recurse_right(root.left, arr, (level + 1))

		return max(right_level, left_level)


	def isSymmetric(self, root: Node):
		if not root:
			return True
		left, right = deque(), deque()
		if root.left and not root.right:
			return False
		if not root.left and root.right:
			return False
		level1 = self.recurse_left(root.left, left, 0)
		level2 = self.recurse_right(root.right, right, 0)

		if level1 != level2:
			return False

		while left and right:
			(nl, l1), (nr, l2) = left.pop(), right.pop()
			if l1 != l2 or  (nl and not nr) or (not nl and nr):
				return False
				
		
		return not left and not right

```

```python
'''
Method 2

Given the root of a binary tree, check if the binary tree has a symmetric structure or not, i.e., left and right subtree mirror each other.

Input:
		   1
		 /   \
		/	  \
	   2	   3
		\	  /
		 \	 /
		  4	5

Output: True

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  / \	  / \
	 /   \	 /	 \
	7	  8	5	  6

Output: True

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  /		  /
	 /		 /
	7		5

Output: False

'''


class Solution:
	def recurse(self, root1, root2):
		if root1 is None and root2 is None:
			return True

  
		if root1 and root2 and root2.data == root2.data:
			return self.recurse(root1.left, root2.right) and self.recurse(root1.right, root2.left)
  
		return False

	def isSymmetric(self, root: Node):
		if not root:
			return False

		return self.recurse(root, root)

```

```python
'''

Given the root of a binary tree, check if it is a complete binary tree or not. A complete binary tree is a binary tree in which every level, except possibly the last, is filled, and all nodes are as far left as possible.

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  / \	  /
	 /	 \	 /
	4	 5	6

Output: True

Input:
		   1
		 /   \
		/	  \
	   2	   3
	  /		  / \
	 /		 /	 \
	4		5	  6

Output: False

'''



class Solution:

    def isComplete(self, root: Node) -> bool:
        if not root:
            return True

        q = deque()
        q.append((root, 0))
        levels = dict()
        while q:
            node, level = q.popleft()
            if level in levels and levels[level] is False:
                return False
            if not node:
                levels[level] = False
                continue

            
            if level not in levels:
                levels[level] = True    
            if level - 1 in levels and levels[level - 1] is False and levels[level]:
                return False
            if not node.left and not node.right:
                levels[level + 1] = False
                continue
            q.append((node.left, level + 1))
            q.append((node.right, level + 1))
        
        return True

            
            
            
```

```python
'''

Given the root of two binary trees, x and y, determine whether y is a subtree of x. A subtree of a tree T is a tree consisting of a node in T and all of its descendants in T.

Input:
		   1						3
		 /   \					  /   \
		/	  \					 /	   \
	   2	   3				6		7
	  /	\	  / \
	 /	 \ 	 /	 \
	4	  5	6	  7

Output: True
Explanation: y is a subtree of x as y = x.right.

Input:
		   1						2
		 /   \					  /   \
		/	  \					 /	   \
	   2	   3				4		5
	  /	\	  / \			   /
	 /	 \ 	 /	 \			  /
	4	  5	6	  7			 8

Output: False

'''



class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''
    
    def recurse(self, x, y, is_root):
        if not x and not y:
            return True
        if is_root and x and x.data == y.data:
            return True
        if (x and not y) or (not x and y):
            return False
        if x.data == y.data:
            flag = self.recurse(x.left, y.left, is_root) and self.recurse(x.right, y.right, is_root)
            if flag:
                return True
        return self.recurse(x.left, y, is_root) or self.recurse(x.right, y, is_root)

    def isSubtree(self, x: Node, y: Node) -> bool:
        if not x:
            return not bool(y)
        if x and not y:
            return True
        
        if x and y and x.data == y.data and not y.left and not y.right:
            return True
        return self.recurse(x, y, is_root=(not y.left) and (not y.right))
        
```

```python
'''

Given the root of a binary tree, check if it is a min-heap or not. In order words, the binary tree must be a complete binary tree where each node has a higher value than its parent's value.

Input:
		   2
		 /   \
		/	  \
	   3	   4
	  / \	  / \
	 /	 \	 /	 \
	5	  6	8	  10

Output: True

Input:
		   5
		 /   \
		/	  \
	   3	   8
	  / \	  / \
	 /	 \	 /	 \
	2	  4	6	  10

Output: False

'''


class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def isMinHeap(self, root: Node) -> bool:
        if not root:
            return True
        
        q = deque()
        q.append((root, 0))
        nodes = [root]
        while q:
            node, level = q.popleft()
            if not node:
                continue
            
            if (node.left and node.left.data < node.data) or (node.right and node.right.data < node.data):
                return False
            q.append((node.left, level + 1))
            q.append((node.right, level + 1))
            nodes.append(node.left)
            if nodes[-1] and nodes[-2] is None:
                return False
            nodes.append(node.right)
            if nodes[-1] and nodes[-2] is None:
                return False
        return True

```

```python
'''

Given the root of a binary tree, return the diameter of binary tree. A binary tree diameter equals the total number of nodes on the longest path between any two leaves in it.

Input:

	   1
	 /   \
	/	  \
   2	   3
	\	  / \
	 \	 /	 \
	  4	5	  6
	   / \
	  /	  \
	 7	   8

Output: 6

Explanation: The nodes involved in binary tree's diameter are [4, 2, 1, 3, 5, 7]. Note that the diameter passes through the root node.


Input:

		1
		 \
		  \
		   2
		  / \
		 /	 \
		3	  4
	   / \	   \
	  /	  \		\
	 5	   6	 7

Output: 5

Explanation: The nodes involved in binary tree's diameter are [5, 3, 2, 4, 7]. Note that the diameter does not pass through the root node.

'''



class Solution:

    '''
    A binary tree node is defined as:

    class Node:
        def __init__(self, data=None, left=None, right=None):
            self.data = data	# data field
            self.left = left	# pointer to the left child
            self.right = right	# pointer to the right child
    '''

    def recurse_diameter(self, root, leaves):
        if not root:
            leaves[root] = 0
            return 0
        leaves[root] = 1
        
        left_diam = self.recurse_diameter(root.left, leaves)
        leaves[root] = max(leaves[root], leaves[root.left] + 1)
        
        right_diam = self.recurse_diameter(root.right, leaves)
        leaves[root] = max(leaves[root], leaves[root.right] + 1)
            
        return max(left_diam, right_diam, leaves[root.left] + leaves[root.right] + 1) 
            
            

    def findDiameter(self, root: Node) -> int:
        if not root:
            return 0
        leaves = dict()
        return self.recurse_diameter(root, leaves)

```
